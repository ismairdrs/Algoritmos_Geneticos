```
#algoritmo genético -> problema da senha

import random
POPULACAO_TAM = 100
GENES = '''abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ 1234567890 ,.-:_!"#%¨&/()=?@${[]}'''
ALVO = "Eu amo estudar"
```

```
class Cromossomo(object):
    def __init__(self, cromossomo):
        self.cromossomo = cromossomo
        self.fitness = self.cal_fitness()
    
    @classmethod
    def mutar_genes(self):
        global GENES
        gene = random.choice(GENES)
        return gene

    @classmethod
    def criar_gnomo(self):
        global ALVO
        gnomo_len = len(ALVO)
        return [self.mutar_genes() for _ in range(gnomo_len)]
    
    def reproduzir(self, par2):
        cromossomo_filho = []
        for gp1, gp2 in zip(self.cromossomo, par2.cromossomo):
            prob = random.random()
            if prob < 0.45: cromossomo_filho.append(gp1)
            elif prob < 0.90: cromossomo_filho.append(gp2)
            else: cromossomo_filho.append(self.mutar_genes())
        return Cromossomo(cromossomo_filho)

    def cal_fitness(self):
        global ALVO
        fitness = 0
        for gs, gt in zip(self.cromossomo, ALVO):
            if gs != gt: fitness+=1
        return fitness
```

```
def main():
    global POPULACAO_TAM
    geracao = 1
    encontrado = False
    populacao = []
    for _ in range(POPULACAO_TAM):
        gnomo = Cromossomo.criar_gnomo()
        populacao.append(Cromossomo(gnomo))
    while not encontrado:
        populacao = sorted(populacao, key = lambda x: x.fitness)
        if populacao[0].fitness <= 0:
            print("Geração: {}\tString: {}\tFitness: {}".format(geracao, "".join(populacao[0].cromossomo), populacao[0].fitness))
            encontrado = True
            break
        geracao_nova = []
        s = int((10*POPULACAO_TAM) / 100)
        geracao_nova.extend(populacao[:s])
        s = int((90*POPULACAO_TAM)/100)
        for _ in range(s):
            pai = random.choice(populacao[:50])
            mae = random.choice(populacao[:50])
            filho = pai.reproduzir(mae)
            geracao_nova.append(filho)
        populacao = geracao_nova
        print("Geração: {}\tString: {}\Fitness: {}".format(geracao, "".join(populacao[0].cromossomo), populacao[0].fitness))
        geracao +=1
        print("Geração: {}\tString: {}\tFitness: {}".format(geracao, "".join(populacao[0].cromossomo), populacao[0].fitness))
```

```
main()
```
